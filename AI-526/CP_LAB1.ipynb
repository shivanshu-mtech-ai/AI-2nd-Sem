```python
from numba import cuda
```

```python
import numpy as np
```

```python
@cuda.jit
def hello_kernel():
  tx=cuda.threadIdx.x # threads id
  bx=cuda.blockIdx.x # block id
  bdim=cuda.blockDim.x # block width

  gid=tx+bx*bdim # global id
  print("hello from bolck",bx,"threads",tx,"GlobalId",gid)


```

```python
blocks=100
threads=400
hello_kernel[blocks,threads]()
cuda.synchronize()
```
